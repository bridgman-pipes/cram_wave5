```python
# Defining parameters for the plot
params = {'axes.titlesize' : 10,
        'axes.labelsize' : 8,
        'lines.linewidth' : 2,
        'lines.markersize' : 3,
        'xtick.labelsize' : 8,
        'ytick.labelsize' : 8,
         'legend.fontsize' : 8,
         'font.weight' : 'normal'}
```

```python
# Mean proportion of individual income
x = np.arange(15)
col = []
for val in x:
    if val<=4:
        col.append('blue')
    elif val >=10:
        col.append('green')
    else:
        col.append('red')

plt.rcParams.update(params)

overall_mean = [0.47, 0.38, 0.41, 0.39, 0.35, 0.23, 0.16, 0.18, 0.17, 0.17, 0.15, 0.12, 0.16, 0.14, 0.15] 
overall_CI = [0.013829368, 0.0142198, 0.012329576, 0.013493228, 0.013107108, 0.01166298, 0.010852716, 0.009735908, 0.010471104, 0.01019984, 0.00932568, 0.010771768, 0.01273804, 0.011044012, 0.012651212]

width = 0.7

plt.bar(x, overall_mean, width, color = col)
plt.errorbar(x,overall_mean, yerr=overall_CI, color='black', fmt='o')

plt.xticks(x, ['April', 'June', 'October', 'January', 'March', 'May/June', 'July/Aug', 'Nov/Dec', 'Feb/March', 'April/May', 'May/June', 'July/Aug', 'Nov/Dec', 'Feb/March', 'April/May'], rotation=45
)


#plt.xlabel("Month")
plt.ylabel("Proportion of Households")
#plt.legend(["Females", "Males"], title = 'Gender', title_fontsize = 8)
plt.title("Food Insecurity and Hunger: NIDS CRAM Wave 5")
plt.figtext(0.00000, 0.00000, "Source: Authors' computation using the NIDS-CRAM Wave 5 data (2021).",
            fontsize=5, ha='left', va='top')
plt.savefig('cram_w5_fig1.png')
plt.show()
```
